## Loading data from API 

```python
import json

# j = json.dumps(json_example)
# json.loads(j)
```

```python
pokemon_name = 'ninetales'
pokemon_url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_name}'
```

```python
pokemon_url
```

```python
import requests

url_content = requests.get(url=pokemon_url)
pokemon_dict = json.loads(url_content.text)
```

```python
data_poke = url_content.json()
```

```python
data_poke
```

```python
print('ninetales abilities:')
for item in data_poke['abilities']:
    print(item['ability']['name'])
```

```python
print('ninetales abilities:')
for index, item in enumerate(data_poke['abilities']):
    print(index, item['ability']['name'])
```

## Json string to dict

```python
example_json_str = '{"activity":{"name":"Cssia","type":"content writing"}}'
```

```python
loaded_str = json.loads(example_json_str)
```

```python
type(loaded_str)
```

```python
loaded_str['activity']['name']
```

## Opening json file

```python
with open('activity_data.json', 'w') as f:
   json.dump(example_json_str, f)
```

```python
with open('activity_data.json') as example_json_file:
   json_file = json.load(example_json_file)
```

```python
type(json_file)
```

```python
converted_json = json.loads(json_file)
```

```python
print(type(converted_json))
print(converted_json)
```
